```python
def stat(df):
    df["datetime"]= pd.to_datetime(df["timestamp"])
    df['dates'] = df['datetime'].dt.date
    t = df.groupby(['DeviceName','dates'], sort=False)['ApplicationBusiness','cpu_usage_percent','mem_usage','disk_r_throughput','disk_w_throughput','network_in_throughput'].max()
    df1 = pd.DataFrame(t.reset_index())
    df1.columns = ['DeviceName','Dates','Area','Max_CPU_Usage','Max_Mem_Usage','Disk_R','Disk_W','Network_In']
    m = df1.groupby(['DeviceName','Area'])['Max_CPU_Usage','Max_Mem_Usage','Disk_R','Disk_W','Network_In'].mean()
    cpum1 = pd.DataFrame(m.reset_index())
    cpum1.columns = ['DeviceName','Area','Mean_Max_CPU_Usage','Mean_Max_Mem_Usage','Mean_Max_DiskR','Mean_Max_DiskW','Mean_Max_Network_In']
    return cpum1
```

```python
refineddf = []
for i in range(1,101):
    tmpdf = pd.read_csv("../csv_data/app_"+str(i)+".csv")
    refineddf.append(stat(tmpdf))
```

```python
refineddf[23]  #Test function 
```

```python
target = pd.concat(refineddf,ignore_index=True)
target
```
